## Step 4. 중동&북아프리카 지역과 남아메리카, 서유럽,남동아시아 총 4개의 지역별로 테러 공격 형태, 사망자와 사상자의 수 등에 대해 각 지역별로 특성들이 있는지를 확인하세요

```python
region_df = pd.read_csv("../data/globalterrorismdb_0617dist.csv",sep=',', usecols=[9,10,19,26,27,28,29,30,31,32,33,34,36,37], encoding = "ISO-8859-1")
```

```python
region_df.tail()
```

```python
region_df.sort_values(by="region",ascending=True)
```

```python
sns.countplot(x="region", data=region_df)
plt.title("지역별 테러수")
plt.show()
```

### 1위 Middle East & North Africa, 2위 South Asia, 3위 South America

```python
sns.countplot(x="attacktype1", data=region_df)
plt.title("공격 타입")
plt.show()
```

### 1위 폭탄 테러, 2위 무장 공격, 3위 암살

```python
sns.countplot(x="region_txt", data=region_df)
plt.title("지역별 테러수")
plt.show()
```

```python
sns.boxplot(x="region", y="attacktype1", data=region_df)
plt.title("지역과 공격타입의 Boxplot")
plt.show()
```

```python
sns.jointplot(x="region", y="attacktype1", data=region_df, kind="kde")
plt.suptitle("지역과 어택타입의 Kernel Density Plot", y=1.02)
plt.show()
```

### 1위 폭탄 테러 3, 2위 무장 공격 2, 3위 암살 1
### 1위 Middle East & North Africa 10, 2위 South Asia 6, 3위 South America 3

### 폭탄 테러로 인한 사상자가 제일 많음
### 무장 테러로 인한 사상자가 2번째
### 암살로 인한 사상자가 3번째

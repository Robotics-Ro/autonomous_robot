```python
import urllib
import re
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import Request

url = "https://search.shopping.naver.com/book/search?bookTabType=BEST_SELLER&catId=50005542&pageIndex=1&pageSize=40&query=%EB%84%A4%EC%9D%B4%EB%B2%84%20%EC%B1%85&sort=REL"

res = urlopen(url)

soup = BeautifulSoup(res, "html.parser")

titles = soup.find_all('span', "bookListItem_text__bglOw") # 책 제목
details = soup.find_all('div', "bookListItem_detail__RBQ6x") # 책 저자, 출판사, 출판일
prices = soup.find_all('span', "bookPrice_price__zr5dh") # 책 가격

goods = []

for n in range(50):
    titles = soup.find_all('span', "bookListItem_text__bglOw")[n].get_text() # 책 제목
    details = soup.find_all('div', "bookListItem_detail__RBQ6x")[n].get_text() # 책 저자, 출판사, 출판일
    prices = soup.find_all('span', "bookPrice_price__zr5dh")[n].get_text() # 책 가격
    goods.append([titles, details, prices])

#for n in range(50):
#    titles = soup.find_all('span', 'class')[n].string
#    goods.append(titles)

goods
```

```python
import pandas as pd

df = pd.read_csv('../data/2020_crime.csv', encoding='euc-kr')
df.head()
```

```python
sql = """CREATE TABLE crime_status
         ( 
            year year,
            police_station varchar(16),
            crime_type varchar(16),
            status_type char(2),
            case_number int
         );"""
cursor = remote.cursor(buffered=True)
cursor.execute(sql)
```

```python
sql = """INSERT INTO crime_status VALUES ("2020",%s, %s, %s, %s)"""
cursor = remote.cursor(buffered=True)
```

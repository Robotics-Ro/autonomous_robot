### 테스트용 데이터 준비

```python
import numpy as np

x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x + 0.5)
y3 = 3 * np.sin(x + 1.0)
y4 = 4 * np.sin(x + 1.5)
```

```python
sns.set_style("white")

plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)

sns.despine(offset=10)

plt.show();
```

### tips data

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show();
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show();
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, color=1)
plt.show();
```

### swarmplot

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=7)
plt.show();
```

```python
tips[tips["day"] == "Sat"].describe()
```

### flights data

```python
flights = sns.load_dataset("flights")
flights.tail(1)
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
flights_pivot = flights.pivot("month", "year", "passengers")
flights_pivot.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```

### Iris data

```python
# sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head()
```

### pairplot

```python
sns.pairplot(iris, hue="species")
plt.show();
```

```python
sns.pairplot(
    iris,
    x_vars=["sepal_width", "sepal_length"],
    y_vars=["petal_width", "petal_length"],
    hue="species"
)

plt.show();
```

### catplot

```python
titanic = sns.load_dataset("titanic")
titanic.tail(2)
```

```python
sns.catplot(
    x="sex",
    y="survived",
    hue="pclass",
    data=titanic,
    kind="bar",
    legend = True,
    hegihts = 7,
    palette = "muted",
)

plt.show()
```

```python
sns.catplot(
    x="deck",
    kind="count",
    data=titanic,
    palette="ch:.25"
)

plt.show()
```

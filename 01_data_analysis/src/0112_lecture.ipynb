### 파이썬의 함수
함수의 정의 : def:  
*arg : 함수선언시 변수를 안넣어도, 무제한으로 넣어도 괜찮다  
**kwarg : dic 형태로 받는다  

````python
def test_arg(a, *arg):
    print('첫번째 인자:',a)
    for each in arg:
        print(f"*arg 다른인자 {each}")

test_arg('a','b')
````

````python
# 인자 **kwarg
def test_kwarg(**kwargs):
    print(kwargs)

test_kwarg(A='aa', B='bb')
````

````python
# 인자 **kwarg 2
def test_kwarg(**kwargs):
    if kwargs is not  None:
        for key, value in kwargs.items():
            print(f'{key} : {value}')

    

test_kwarg(A='aa', B='bb')
````

````python
# 인자 **kwarg 를 활용한 기본값으로 처리
def test_kwargs(**kwargs):
    A = kwargs.get('A',1)
    return(A)


print(f'인자를 안넣을시 : {test_kwargs()}')
print(f'인자를 넣을시 : {test_kwargs(A=10)}')
````

### 예제

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
````

````python
# 일반적인 작성 형태
def plotSinWave(amp, freq, endTime, sampleTime, bias):
    time = np.arange(0, endTime, sampleTime)
    result = amp *np.sin(2*np.pi*freq*time) + bias

    plt.figure(figsize=(12,6))
    plt.plot(time, result); plt.grid(True)
    plt.xlabel('time'); plt.ylabel('sin')
    plt.title(str(amp) +'*sin(2*pi*' + str(freq) + 't )+' +str(bias))
    plt.show()
plotSinWave(2, 1, 10, 0.01, 0)
````

````python
# **kwarg를 사용해서 디폴트 값을 지정한 함수
# amp, freq, endTime, sampleTime, bias)
def plotSinWave(**kwargs):
    endTime, sampleTime = kwargs.get('endTime',1), kwargs.get('sampleTime', 0.01)
    amp, freq, bias = kwargs.get('amp',1), kwargs.get('freq',1), kwargs.get('bias',0)

    time = np.arange(0, endTime, sampleTime)
    result = amp *np.sin(2*np.pi*freq*time) + bias

    figsize = kwargs.get('figsize', (12,6))

    plt.figure(figsize=figsize) # 여기서 우측 figsize 는 옵션명이다
    plt.plot(time, result); plt.grid(True)
    plt.xlabel('time'); plt.ylabel('sin')
    plt.title(str(amp) +'*sin(2*pi*' + str(freq) + 't )+' +str(bias))
    plt.show()
    
# plotSinWave(2, 1, 10, 0.01, 0)
plotSinWave(amp = 2, figsize = (12,5))
````

위에서 적은 self는 클래스내에서 tmp 로 대체됨을 알 수 있다

````python
class DrawSin:
    def __init__(self, amp, freq, bias, end_time):
        self.amp = amp
        self.freq = freq
        self.bias = bias
        self.end_time = end_time
    
    def calc_sin(self):
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp * np.sin(2*np.pi*self.freq*self.t) + self.bias

    
    def draw_sin(self):
        y = self.calc_sin()
        plt.figure(figsize=(12,6))
        plt.plot(self.t,y)
        plt.grid()
        plt.show()


ds = DrawSin(1,1,0,3)
ds.draw_sin()
````

````python
# 아래와 같이 클래스내 객체를 변경 할 수도 있다
ds.freq = 2
ds.draw_sin()
````

### 클래스 상속에 관하여

````python
class DrawSinusoidal(DrawSin):
    def calc_cos(self):
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp * np.cos(2*np.pi*self.freq*self.t) + self.bias

    def draw_cos(self):
        y = self.calc_cos()
        plt.figure(figsize=(12,6))
        plt.plot(self.t, y)
        plt.grid()
        plt.show()
````

### 오버라이딩 및 super
아래는 기존(DrawSinusoidal)에 있던 draw_sin를 다시 불러오는게 아니라 덮어씌우는(오버라이딩)이 됐다

### super
만약 오버라이딩이 아닌, 추가를 하고 싶으면 어떻게 해야할까?

````python
class DrawSinusoidal3(DrawSinusoidal2):
    def __init__(self, amp, freq, bias, end_time, ts):
        self.ts = ts
````

````python
ds3 = DrawSinusoidal3(1,1,0,3,0.1)
ds3.__dict__
````

원래는 ts(간격)을 ```__init__```에 추가하고 싶었지만 오버라이딩 되어버렸다.

````python
class DrawSinusoidal3(DrawSinusoidal2):
    def __init__(self, amp, freq, bias, end_time, ts):
        super().__init__(amp, freq, bias, end_time)
        self.ts = ts # 여기가 추가

    def calc_sin(self): #여기는 오버라이딩
        self.t = np.arange(0,self.end_time, self.ts)
        return self.amp * np.sin(2*np.pi*self.freq*self.t) + self.bias

    def draw_sin(self): #여기도 오버라이딩
        y = self.calc_sin()
        plt.figure(figsize=(12,6))
        plt.plot(self.t, y)
        plt.title('Sin Graph')
        plt.xlabel('time(sec)')
        plt.grid()
        plt.show()

    
````

````python
ds3=DrawSinusoidal3(1,1,0,3,0.1)
ds3.draw_sin()
````
